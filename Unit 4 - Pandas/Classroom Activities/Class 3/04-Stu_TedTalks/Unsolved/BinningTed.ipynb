```python
# Import Dependencies
import pandas as pd
```

```python
# Create a path to the csv and read it into a Pandas DataFrame
csv_path = "Resources/ted_talks.csv"
ted_df = pd.read_csv(csv_path)

ted_df.head()
```

```python
# Figure out the minimum and maximum views for a TED Talk
min = ted_df["views"].min()
max = ted_df["views"].max()
print(max)
min
```

```python
# Create bins in which to place values based upon TED Talk views
bins = []
for x in range(ted_df["views"].min(), ted_df["views"].max() + 4717667, 4717667):
    bins.append(x)
    
#print(bins)
# Create labels for these bins
labels = []
for item in range(len(bins) - 1):
    label = str(bins[item]) + " to " + str(bins[item + 1])
    #print(label)
    labels.append(label)
#print(labels)
print(len(bins))
print(labels)
```

```python
# Slice the data and place it into bins
ted_df["View Group"] = pd.cut(ted_df["views"], bins, labels = labels)
ted_df.head()
```
