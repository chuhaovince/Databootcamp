```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
Temp_F = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
month = np.arange(1,13,1)
```

```python
plt.scatter(month, Temp_F)
plt.show()
```

```python
Temp_C = [(x-32)*0.56 for x in Temp_F]
```

```python
plt.scatter(month, Temp_C)
```

```python
plt.plot(month, Temp_F)
plt.plot(month, Temp_C)
plt.show()
```
