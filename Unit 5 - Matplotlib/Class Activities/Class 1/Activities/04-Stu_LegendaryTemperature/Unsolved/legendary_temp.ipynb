```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axis = np.arange(1,13,1)
```

```python
F = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
C = [(x-32) * 0.56 for x in F]
```

```python
fahrenheit, = plt.plot(x_axis, F, marker="p", color = "blue", linewidth = 3, label = "Fahrenheit")
celcius, = plt.plot(x_axis, C, marker="^", color = "red", linewidth = 1, label = "Celcius")
```

```python
plt.legend(handles = [fahrenheit, celcius], loc = "best")
```
