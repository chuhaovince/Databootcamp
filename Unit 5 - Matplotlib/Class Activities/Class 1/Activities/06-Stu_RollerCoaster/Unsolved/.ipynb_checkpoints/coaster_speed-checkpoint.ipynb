```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
danger_drop = [9,8,90,85,80,70,70,65,55,60,70,65,50]
railgun = [75,70,60,65,60,45,55,50,40,40,35,35,30]
```

```python
time = np.arange(0,121, 10)
time
```

```python
plt.plot(time, danger_drop, label = "Danger Drop", color = "red", marker = "x")
plt.plot(time, railgun, label = "RailGun", color = "blue", marker = "o")
plt.title("Coaster Speed Over Time")
plt.xlabel("Coaster Runtime")
plt.ylabel("Speed(MPH)")
plt.legend(loc = "top right") # loc = "best"
plt.xlim(0,120)
plt.ylim(0,95)
plt.grid()
```
