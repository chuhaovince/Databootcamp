```python
# Dependencies
import requests
from config import api_key

url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Store a search term
query = "Trudeau"

# Search for articles published between a begin and end date
begin = 20120101
end = 20130101

# Build url
query_url = f"api-key={api_key}&q={query}&begin_date={begin}&end_date={end}"
```

```python
# Retrieve articles
response = requests.get(url+query_url).json()
response["response"]["docs"]
```

```python
for i in response["response"]["docs"]:
    print(f'A snippet from the article: {i["snippet"]}')
    print(f'============================================')
```
